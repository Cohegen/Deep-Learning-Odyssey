## 1.1 The Test Set

* Suppose we have a classifier $f$ and that we have a fresh datasets $D = (x^{(i)},y^{(i)})$ which has not been used to train the classifier $f$.
* The **empirical error** of our classifier $f$ on $D$ is simply the fraction of instances for which the prediction $f(x^{(i)})$ disagree with the true $y^{(i)}$ in a finite dataset(test set), and is given the following expression:
  $\epsilon_D(f) = \frac{1}{n} \sum_{i=1}^{n} \mathbb{1}(f(\mathbf{x}^{(i)}) \neq y^{(i)})$.

* However, the **population error** is the **expected** fraction (probability) of examples in the underlying population with some distribution $P(X,Y)$ characterized by the probability density function $p(x,y)$ for which our classifier disagrees with the true label.
* Because the population error cannot be observed directly,the test error serves as a statistical estimator of it.
* Because each prediction is a Bernoulli random variable, the test error is the sample mean of i.i.d(independent indentically distributed) random variables and thus an unbiased estimator of the popolation error.
* By the central limit theorem, this estimate converges to the true error at the rate of $O(1/\sqrt{n})$, meaning that improving precision requires larger test sets.
* The variance of the estimate depends on the true error rate and is maximized near 0.5, giving a worst-case bound on uncertainity.
* As a result, estimating error within $+/- 0.01$ with high confidence requires thousands to tens of thousands of test samples.
